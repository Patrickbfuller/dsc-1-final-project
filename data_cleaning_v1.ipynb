To do: bathrooms to int. Floors to int. Waterfront to bool. 
View to int (NAN values dropped). sqft_basement to int. zipcode to str. 

```python
dd.bathrooms = dd.bathrooms.astype(int)
```

```python
dd.floors = dd.floors.astype(int)
```

```python
dd.waterfront.fillna(0, inplace = True)
```

```python
dd.waterfront = dd.waterfront.astype(bool)
```

```python
dd = dd.dropna(subset = ["view"])
```

```python
dd.view.unique()
```

```python
dd.info()
```

```python
dd.zipcode = dd.zipcode.astype(str)
```

```python
dd.info()
```

```python
dd.sqft_basement.unique()
```

```python
dd.sqft_basement = dd.sqft_basement.astype(int)
```

```python
dd.to_csv("data_cleaned_v1")
```
